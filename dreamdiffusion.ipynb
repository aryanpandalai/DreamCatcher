```python
import torch
from diffusers import LTXPipeline
from diffusers.utils import export_to_video
```

```python
pipe = LTXImageToVideoPipeline.from_pretrained("Lightricks/LTX-Video", torch_dtype=torch.bfloat16)
pipe.to("cuda")

image = load_image(
    "/content/horse.png"
)
prompt = (
    "A strong, muscular chestnut horse gallops through a vast open field, its mane flowing wildly in the wind. "
    "Its hooves pound against the soft earth, kicking up small clouds of dust with every stride. "
    "Horse run and the camera tracks horse"
    "The scene is full of energy, with the camera dynamically following alongside, slightly shaking with each powerful step. "
    "A few birds fly overhead as the wind rustles the tall grass, adding a sense of motion and freedom to the scene. "
)

# Define negative prompt to avoid animation artifacts
negative_prompt = (
    "worst quality, blurry, pixelated, glitchy, unnatural motion, stiff animation, "
    "overexposed, cartoonish, floating mane, rigid leg movement, frozen horse mid-run, "
    "inconsistent camera tracking, unnatural lighting, color banding, jittery motion."
)
video = pipe(
    image=image,
    prompt=prompt,
    negative_prompt=negative_prompt,
    width=480,
    height=704,
    num_frames=161,
    num_inference_steps=100,
).frames[0]
export_to_video(video, "horse.mp4", fps=24)
```
